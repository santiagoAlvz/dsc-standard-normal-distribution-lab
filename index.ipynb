## Load, visualize and give general comments about the dataset

## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
mean = float(df.mean())
sd = float(df.std())
print(f'Min. boundary: {mean - 3*sd}, Max. boundary: {mean + 3*sd}')
```

```python
# Write your answer here 
## 99% of the apple orchard observations lie within 3 standard deviations from the mean, more specifically, 99% of all recorded yields
## weight between 24.39 lbs and 60.41 lbs
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
z = (35 - mean) / sd
z
```

```python
# Interpret the result
# Having a yield of 35 pounds in this distribution is equivalent to having -1.23 in a standard normal distribution. The yield is
# considerably below average, outside of the 68% most common vaues
```

## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

```python
# Interpret the z score
# This tree has a yield 1.85 standard deviations above the mean, meaning it has better yields than ~99% of the trees. 
```

```python
# Calculate yield
yd = (1.85 * sd) + mean
```

```python
# What is the yield ?
yd
```

##  Convert each tree’s yield to a z-score so the new variable is the “z-score for weight”

The units are still the apple trees. For the data set of all z-scores:

* What is the shape? 
* The mean? 
* The standard deviation?

```python
# Give your solution here 
import numpy as np

zscores = (df - mean) / sd
zmean = float(zscores.mean())
zsd = float(zscores.std())

zscores.plot.hist(bins=30)
print(f'Mean: {zmean}, SD: {zsd}')
```

```python
# Your observations
# The mean for the converted distribution is a very small value, practically 0 when rounded. The standard deviation, meanwhile, is 1
# Applying the z-score to the whole dataset converted this distribution in a standard normal distrbution.
```
